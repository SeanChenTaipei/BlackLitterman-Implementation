### generate技術指標，並存檔

```python
ta_list = talib.get_functions()
for filename in file:
    df = pd.read_csv('./re_data/%s'%filename,encoding = 'utf-8')
    data = df[['date','open','high','low','close','volume']]
    # 迴圈執行，看看結果吧！
    for index in ta_list:
        try:
            # index 為技術指標的代碼，透過迴圈填入，再透過 eval 計算出 output
            output = eval('abstract.'+index+'(data)')
            # 如果輸出是一維資料，幫這個指標取名為 index 本身；多維資料則不需命名
            output.name = index if type(output) == pd.core.series.Series else None
            # 透過 merge 把輸出結果併入 df DataFrame
            df = pd.concat([df, pd.DataFrame(output)],axis=1)
        except:
            print(index+" failed")
    df.fillna(method='bfill',inplace=True)
    df.drop(['COSH', 'EXP','SINH','ACOS','ASIN'], axis=1,inplace=True) #有inf現象所以要刪除
    df.to_csv('./tech_index/tech_%s'%filename[3:],index = False,encoding = 'utf-8-sig')
```
